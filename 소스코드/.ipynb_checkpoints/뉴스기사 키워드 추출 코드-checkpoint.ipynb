```python
!pip install konlpy
from konlpy.tag import Komoran
import requests
import lxml.html
import cssselect as external_cssselect
from lxml.cssselect import CSSSelector
```

```python
url = 'https://search.daum.net/search?nil_suggest=btn&w=news&DA=STC&cluster=y&q=산불+원인&period=y&sd=20180804154732&ed=20190804154732'
url
```

```python
urls = [] 
for page in range(1, 500): 
    res = requests.get(url.format(page)) 
    root = lxml.html.fromstring(res.text) 
    for link in root.cssselect('a.f_nb'): 
         urls.append(link.attrib['href']) 
```

```python
len(urls)
```

```python
!pip install tqdm
```

```python
import tqdm
```

```python
articles = [] 
i=0
for u in tqdm.tqdm_notebook(urls): 
    if not u.startswith('http'): 
        continue
    res = requests.get(u) 
    print(i ,' ',res.status_code)
    i = i+1
    root = lxml.html.fromstring(res.text) 
    body = root.cssselect('.article_view').pop() 
    content = body.text_content().strip() # 본문을 가져와 앞뒤 공백을 제거
    articles.append(content) 
```

```python
import pandas
df = pandas.DataFrame({'article': articles}) 
df.to_excel('daum_news_ai.xlsx') 
```

```python
from konlpy.tag import Twitter 
```

```python
tagger = Twitter()
```

```python
def kor_noun(text): 
    words = [] 
    for w in tagger.nouns(text): 
        if len(w) > 1: 
             words.append(w) 
    return words 
```

```python
kor_noun('한글은 한국어를 표현하는 문자인 것이다.') 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(tokenizer=kor_noun, max_features=1000)
```

```python
articles = df['article'].astype(str)
```

```python
tdm = cv.fit_transform(articles)
```

```python
words = cv.get_feature_names()
```

```python
sorted(zip(tdm.sum(axis=0).flat, words), reverse=True)[:20]
```

```python
!pip install gensim 
```

```python
from gensim.matutils import Sparse2Corpus 
corpus = Sparse2Corpus(tdm.T)
```

```python
from gensim.models.ldamodel import LdaModel 
lda = LdaModel(corpus=corpus, id2word=dict(enumerate(words)), 
 num_topics=10, passes=20, iterations=50, 
 random_state=1234) 
```

```python
lda.show_topic(0)
```

```python
lda.show_topic(1) 
```

```python
def tdm2doc(tdm, i): 
    doc = [] 
    for i, n in enumerate(tdm[i].toarray().flat): 
        if n > 0: 
             doc.append((i, n)) 
    return doc
```

```python
doc = tdm2doc(tdm, 50)
```

```python
lda.get_document_topics(doc)
```
