```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
import pandas as pd

df = pd.DataFrame(data=[],columns=['href'])
display(df)
```

```python
topics=['politics','economy','social','culture','education','science','history','philosophy','sports','environment']
page =[2,2,4,1,2,2,1,1,1,1,1]

cnt=0
topics_index = 0

for topics_index in range (0,len(topics)):
    for i in range(1,page[topics_index]+1):
        if i==1:
            url = "https://debatingday.com/category/opendiscussion/"+topics[topics_index]
        else :
            url = "https://debatingday.com/category/opendiscussion/"+topics[topics_index]+"/page/"+str(i)+"/"

        html = urlopen(url)
        source = html.read() # 바이트코드 type으로 소스를 읽는다.
        html.close() # urlopen을 진행한 후에는 close를 한다.

        soup = BeautifulSoup(source, "html5lib") # 파싱할 문서를 BeautifulSoup 클래스의 생성자에 넘겨주어 문서 개체를 생성, 관습적으로 soup 이라 부름
        links = soup.find_all(class_="blog-one-item")
        for link in links:
            cnt+=1
            link_url = link.a.get('href')
            print(cnt,link_url)
            a = pd.DataFrame(data=[link_url],columns=['href'])
            df = df.append(a)
            df = df.reset_index(drop=True)
```

```python
display(df)
```
