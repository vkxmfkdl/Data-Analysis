```python
import requests
```

```python
# 파이썬 홈페이지에서 로고 파일을 받기.
url = 'https://www.python.org/static/img/python-logo.png'
res = requests.get(url)
```

```python
# http 상태코드를 확인
# 200인것을 보니 잘 받아진거로 확인.
res.status_code
```

```python
#.content 에 응답내용이 담겨있다.
# 이를 저장하면 그림파일이 된다.
with open('logo.png','wb') as f:
    f.write(res.content)
```

```python
url = 'http://www.gutenberg.org/files/2591/2591-0.txt'
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
# res.text는 res.content의 내용이 텍스트로 해석된 것 이다.
res.text[:100]
```

```python
#단어 추출 , 알파벳으로 된 단어를 추출한다.
import re
```

```python
words = re.findall(r'[a-zA-Z]+',res.text)
```

```python
words[:10]
```

```python
#각 단어들이 몇 번 씩 등장했는지 세어보자.
word_count ={}
for word in words:
    word = word.lower()
    if word in word_count:
        word_count[word]+=1
    else :
        word_count[word]=1
```

```python
word_count
```

```python
from collections import Counter
```

```python
#Counter를 사용하면 좀 더 간단히 할 수 있음.
#Counter는 사전과 동일하지만 모든 키에 대해 기본값으로 0을 가지고 있어서
#단어가 있는지 확인할 필요가 없다.

word_count= Counter()
for word in words:
    word = word.lower()
    word_count[word]+=1
```

```python
word_count
```

```python
#pandas를 이용해서 단어빈도수에 따른 정렬을 해보겠다.
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(word_count.items()),columns=['word','n'])
```
