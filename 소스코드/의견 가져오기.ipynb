```python
cnt=0
for url in link_df['href']:
    print("cnt:",cnt)
    time.sleep(5)
    html = urlopen(url)
    source = html.read() # 바이트코드 type으로 소스를 읽는다.
    html.close() # urlopen을 진행한 후에는 close를 한다.

    soup = BeautifulSoup(source, 'html5lib') # 파싱할 문서를 BeautifulSoup 클래스의 생성자에 넘겨주어 문서 개체를 생성, 관습적으로 soup 이라 부름
    opinions = soup.find_all(class_="chat-box chat-in")

    for opinion in opinions:
        opinion = soup.find_all(class_="message flagApp")
        for texts in opinion:
            contents = texts.find_all('p')
            for content in contents:
                list = []
                str = content.text
                list = kkma.sentences(str)
                for item in list:
                    a = pd.DataFrame(data={'text':[item],'label':[0]},columns=['text','label'])
                    df = df.append(a)
    cnt+=1
df = df.reset_index(drop=True)
display(df)


        
```

```python
cnt=0
for url in link_df['href']:
    print("cnt:",cnt)
    time.sleep(5)
    html = urlopen(url)
    source = html.read() # 바이트코드 type으로 소스를 읽는다.
    html.close() # urlopen을 진행한 후에는 close를 한다.

    soup = BeautifulSoup(source, 'html5lib') # 파싱할 문서를 BeautifulSoup 클래스의 생성자에 넘겨주어 문서 개체를 생성, 관습적으로 soup 이라 부름
    opinions = soup.find_all(class_="chat-box chat-in")

    for opinion in opinions:
        opinion = soup.find_all(class_="message flagApp")
        for texts in opinion:
            contents = texts.find_all('p')
            for content in contents:
                list = []
                str = content.text
                list = kkma.sentences(str)
                for item in list:
                    if item=='\xa0':
                        continue
                    a = pd.DataFrame(data={'text':[item],'label':[0]},columns=['text','label'])
                    df = df.append(a)
    cnt+=1
    if cnt==1:
        break
df = df.reset_index(drop=True)
display(df)


        
```
